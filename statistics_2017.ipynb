```python
def loadMatches(filename):
    matches = pd.read_json('data/matches/'+filename)
    match_data = []
    for i in matches["matches"]: 
        record = []
        record.append({"player": i[0]["player"], "score": i[0]["score"]})
        record.append({"player": i[1]["player"], "score": i[1]["score"]})
        match_data.append(record)
    return match_data

def loadAllMatchesFromYear(number_of_drafts, yy):
    match_data = []
    for i in range(1, number_of_drafts+1):
        tmp = loadMatches('matches_'+str(yy)+'_'+str(i)+'.json')
        for j in tmp:
            match_data.append(j)
    return match_data
        
all_matches = loadAllMatchesFromYear(13, 18)
print(len(all_matches))

def extract_players(my_matches):
    players = []
    for i in my_matches:
        players.append(i[0]["player"])
        players.append(i[1]["player"])
    return list(set(players))

def getWinRatioForPlayer(player, matches):
    player_matches = []
    won, lost, m_won, m_lost, total, m_total = 0, 0, 0, 0, 0, 0 
    for i in matches:
        if i[0]['player'] == player or i[1]['player'] == player:
            player_matches.append(i)
    m_total = len(player_matches)
    
    for i in player_matches:
        if i[0]['player'] == player:
            won += i[0]['score']
            lost += i[1]['score']
            if i[0]['score'] > i[1]['score']: m_won +=1
            else: m_lost +=1
        if i[1]['player'] == player:
            won += i[1]['score']
            lost += i[0]['score']
            if i[1]['score'] > i[0]['score']: m_won +=1
            else: m_lost +=1
    total = won+lost
    if total !=0:
        win_rate_games = round(100*(won+0.0)/total, 2)
        win_rate_matches = round(100*(m_won+0.0)/m_total, 2)
        alpha_factor = round(win_rate_matches - win_rate_games, 2)
    else:
        win_rate_games = 0
        win_rate_matches = 0
        alpha_factor = 0
    
    return {'player': player,
           'won': won,
           'lost': lost,
           'matches_won': m_won,
           'matches_lost': m_lost,
           'total': total,
           'matches_total': m_total,
           'win_rate_games':win_rate_games,
           'win_rate_matches':win_rate_matches, 
           'alpha_factor':alpha_factor }
          
all_players = extract_players(all_matches)

for player in all_players:
    print(getWinRatioForPlayer(player, all_matches))
```

```python
# load matches and decklists for specific draft

class deck:
    def __init__(self, player, player_stats, draft, cards, name):
        self.player = player
        self.player_stats = player_stats
        self.name = name
        self.draft = draft
        self.cards = cards
        tmp = name.split(' ')
        self.colors = tmp[0]
        self.type = tmp[1] 

def loadDrafts(filename):
    draft_data = pd.read_json('data/decks/'+filename)
    return draft_data

def loadMatchData(filename):
    match_data = pd.read_json('data/matches/'+filename)
    return match_data

def loadDraftData(draft_number, yy):
    filename_matches = 'matches_'+str(yy)+'_'+str(draft_number)+'.json'
    filename_decks = 'decklists_'+str(yy)+'_'+str(draft_number)+'.json'
    match_data = loadMatchData(filename_matches)
    deck_data = loadDrafts(filename_decks)
    decks = []
    
    for i in range(0, len(deck_data)):
        decks.append(deck(deck_data['player'][i], getWinRatioForPlayer(deck_data['player'][i], match_data['matches']),
                    deck_data['draft'][i], deck_data['cards'][i], deck_data['name'][i]) )
        
    return decks
    
def loadAllDraftDataYear(number_of_drafts, yy):
    all_decks = []
    for i in range(1, number_of_drafts+1):
        tmp = loadDraftData(i, yy)
        for j in tmp:
            all_decks.append(j)
    return all_decks
    
all_decks = loadAllDraftDataYear(13, 18)
    
```

```python
# Calculate colors and archetypes win rates

def getColorsDict(all_decks):
    colors_dict = {}
    for i in all_decks:
        if colors_dict.has_key(i.colors):
            colors_dict[i.colors] +=1
        else:
            colors_dict[i.colors] =1
    return colors_dict

def getArchetypesDict(all_decks):
    archetypes_dict = {}
    for i in all_decks:
        if archetypes_dict.has_key(i.type):
            archetypes_dict[i.type] +=1
        else:
            archetypes_dict[i.type] =1 
    return archetypes_dict

def getArchetypeWinRate(decks, archetype):
    won, lost, m_won, m_lost, total, m_total = 0, 0, 0, 0, 0, 0
    for i in decks:
        if (i.type == archetype):
            won +=i.player_stats['won']
            lost +=i.player_stats['lost']
            m_won +=i.player_stats['matches_won']
            m_lost +=i.player_stats['matches_lost']
            total +=i.player_stats['total']
            m_total +=i.player_stats['matches_total']
            
    win_rate_games = round(100*(won+0.0)/total, 2)
    win_rate_matches = round(100*(m_won+0.0)/m_total, 2)
    alpha_factor = round(win_rate_matches - win_rate_games, 4)
    return {'type': archetype,
           'won': won,
           'lost': lost,
           'matches_won': m_won,
           'matches_lost': m_lost,
           'total': total,
           'matches_total': m_total,
           'win_rate_games':win_rate_games,
           'win_rate_matches': win_rate_matches, 
           'alpha_factor':alpha_factor }
    
def getArchetypesWinRate(decks):
    archetypes = getArchetypesDict(decks)
    for i in archetypes:
        print(i)
        print(getArchetypeWinRate(decks, i))

getArchetypesWinRate(all_decks)

def getColorWinRate(decks, color):
    won, lost, m_won, m_lost, total, m_total = 0, 0, 0, 0, 0, 0
    for i in decks:
        if (i.colors == color):
            won +=i.player_stats['won']
            lost +=i.player_stats['lost']
            m_won +=i.player_stats['matches_won']
            m_lost +=i.player_stats['matches_lost']
            total +=i.player_stats['total']
            m_total +=i.player_stats['matches_total']
            
    win_rate_games = round(100*(won+0.0)/total, 2)
    win_rate_matches = round(100*(m_won+0.0)/m_total, 2)
    alpha_factor = round(win_rate_matches - win_rate_games, 4)
    return {'colors': color,
           'won': won,
           'lost': lost,
           'matches_won': m_won,
           'matches_lost': m_lost,
           'total': total,
           'matches_total': m_total,
           'win_rate_games':win_rate_games,
           'win_rate_matches': win_rate_matches, 
           'alpha_factor':alpha_factor }
    
def getColorsWinRate(decks):
    colors = getColorsDict(decks)
    for i in colors:
        print '\n'+i
        print(getColorWinRate(decks, i))
        
getColorsWinRate(all_decks)
```

```python
# Card win rates

def getCardsDict(decks):
    card_dict = {}
    for i in decks:
        for j in i.cards:
            if card_dict.has_key(j):
                card_dict[j]['won']+=i.player_stats['won']
                card_dict[j]['lost']+=i.player_stats['lost']
                card_dict[j]['matches_won']+=i.player_stats['matches_won']
                card_dict[j]['matches_lost']+=i.player_stats['matches_lost']
                card_dict[j]['total']+=i.player_stats['total']
                card_dict[j]['matches_total']+=i.player_stats['matches_total']
            else:
                card_dict[j] = {'won': i.player_stats['won'], 
                                'lost': i.player_stats['lost'],
                                'matches_won':i.player_stats['matches_won'],
                                'matches_lost':i.player_stats['matches_lost'],
                                'total':i.player_stats['total'],
                                'matches_total':i.player_stats['matches_total']
                               }
                
    cards_games_won = []
    cards_matches_won = []
    for i in card_dict:
        if card_dict[i]['total'] != 0:
            win_rate_games = round(100*(card_dict[i]['won']+0.0)/card_dict[i]['total'], 2)
            win_rate_matches = round(100*(card_dict[i]['matches_won']+0.0)/card_dict[i]['matches_total'], 2)
            alpha_factor = round(win_rate_matches - win_rate_games, 4)
            if card_dict[i]['total'] >=50:
                cards_games_won.append((i, win_rate_games))
                cards_matches_won.append((i, win_rate_matches))
            
    cards_games_won = reversed(sorted(cards_games_won, key=lambda x: x[1]))
    cards_matches_won = reversed(sorted(cards_matches_won, key=lambda x: x[1]))
    
    cards_games_total = []
    for key in card_dict.keys():
        cards_games_total.append([key, card_dict[key]['total']])
        
    cards_games_total = reversed(sorted(cards_games_total, key=lambda x: x[1]))

    return cards_games_won, cards_matches_won, cards_games_total
    
    
card_games_win, card_matches_win, cards_games_total = getCardsDict(all_decks)

for i in card_games_win: print i[0], i[1]
    
for i in cards_games_total: print i[0], i[1]
    
for i in card_matches_win: print i[0], i[1]
```

```python
# Preparing graphs for statistics

import graphviz as gv

import functools
graph = functools.partial(gv.Graph, format='svg')
digraph = functools.partial(gv.Digraph, format='svg')

def add_nodes(graph, nodes):
    for n in nodes:
        if isinstance(n, tuple):
            graph.node(n[0], **n[1])
        else:
            graph.node(n)
    return graph

def add_edges(graph, edges):
    for e in edges:
        if isinstance(e[0], tuple):
            graph.edge(*e[0], **e[1])
        else:
            graph.edge(*e)
    return graph

def apply_styles(graph, styles):
    graph.graph_attr.update(
        ('graph' in styles and styles['graph']) or {}
    )
    graph.node_attr.update(
        ('nodes' in styles and styles['nodes']) or {}
    )
    graph.edge_attr.update(
        ('edges' in styles and styles['edges']) or {}
    )
    return graph

styles = {
    'graph': {
        'label': 'Players and games',
        'fontsize': '16',
        'fontcolor': 'yellow',
        'bgcolor': '#333333',
        'rankdir': 'TB',
    },
    'nodes': {
        'fontname': 'Helvetica',
        'shape': 'oval',
        'fontcolor': 'white',
        'color': 'white',
        'style': 'filled',
        'fillcolor': '#006699',
    },
    'edges': {
        'style': 'solid',
        'color': 'white',
        'arrowhead': 'open',
        'fontname': 'Courier',
        'fontsize': '12',
        'fontcolor': 'white',
    },
}
```

```python
# Use score calculator for the results:

import operator

# input: a set of data with match results
# output: a table with final standings

def extract_players(my_matches):
    players = []
    for i in my_matches:
        players.append(i[0]["player"])
        players.append(i[1]["player"])
    return list(set(players))

def normalized_score(player_score):
    deals = player_score["won"] + player_score["lost"]
    win_rate =(player_score["won"]+0.0)/deals
    return player_score["points"]*100 + round(win_rate *100, 0)

def get_sorted_list(players, player_points):
    standings = {}
    for i in players:
        score = normalized_score(player_points[i])
        standings[i] = score
    sorted_standings = sorted(standings.items(), key=operator.itemgetter(1), reverse=True)

    return sorted_standings

# Loading matches from .json file

def loadMatches(filename):
    matches = pd.read_json('data/matches/'+filename)
    match_data = []
    for i in matches["matches"]: 
        record = []
        record.append({"player": i[0]["player"], "score": i[0]["score"]})
        record.append({"player": i[1]["player"], "score": i[1]["score"]})
        match_data.append(record)
    return match_data

# Calculates score for given set of matches

def calculateScores(match_data):
    players = extract_players(match_data)

    player_points = {}
    for i in players:
        player_points[i] = {"points": 0, "won": 0, "lost": 0}

    for i in players:
        for j in match_data:
            if j[0]["player"] == i:
                player_points[i]["won"] +=j[0]["score"]
                if j[0]["score"] > j[1]["score"]:
                    player_points[i]["points"] += 3
                player_points[i]["lost"] +=j[1]["score"]
            if j[1]["player"] == i:
                player_points[i]["won"] +=j[1]["score"]
                if j[1]["score"] > j[0]["score"]:
                    player_points[i]["points"] += 3
                player_points[i]["lost"] +=j[0]["score"]
    
    standings = get_sorted_list(players, player_points)

    for i in standings:
        won =  player_points[i[0]]['won']
        lost =  player_points[i[0]]['lost']
        win_rate = round((won+0.0)/(won+lost), 2)
        if (won+lost) >= 15:
            print i[0], win_rate
    
    #return standings, player_points

calculateScores(color_matches)

print '   '

calculateScores(archetype_matches)
```
