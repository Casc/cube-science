```python
class RuleDirect():
    def __init__(self, card1, card2, synergy_level):
        self.card1 = card1
        self.card2 = card2
        self.synergy_level = synergy_level
    def showRule(self):
        print(self.card1, self.card2, self.synergy_level)
    def validateDeck(self, deck, cards):
        if self.card1 in deck and self.card2 in deck:
            self.showRule()
        
        
class RuleOneWay():
    def __init__(self, card1, cond2, synergy_level):
        self.card1 = card1
        self.cond2 = cond2
        self.synergy_level = synergy_level
        
    def showRule(self):
        print(self.card1, self.cond2, self.synergy_level)
    
    def validateDeck(self, deck, cards):
        if self.card1 in deck:
            for d in deck:
                if d != self.card1:
                    if list((set(self.cond2) - set(cards[d]))) == []:
                        print(self.card1, d, self.synergy_level)
        
class RuleTwoWay():
    def __init__(self, cond1, cond2, synergy_level):
        self.cond1 = cond1
        self.cond2 = cond2
        self.synergy_level = synergy_level
        
    def showRule(self):
        print(self.cond1, self.cond2, self.synergy_level)
        
    def validateDeck(self, deck, cards):
        for d1 in deck:
            for d2 in deck:
                if d1 != d2:
                    l1 = list((set(self.cond1) - set(cards[d1])))
                    l2 = list((set(self.cond2) - set(cards[d2])))
                    if l1 == l2 == []:
                        print(d1, d2, self.synergy_level)
                    

class synergyChecker():
    def __init__(self, rules, deck, cards):
        self.rules = rules
        self.deck = deck
        self.cards = cards
    
    def validateAllRules(self):
        for r in rules:
            r.validateDeck(self.deck, self.cards)
    
deck = ['Goblin Bombardier', 'Elvish Visionary', 'Braintaster', 'Mogg Fanatic', 'Crystal Shard', 'Island', 
        'Goblin Char', 'Goblin Dragon Hunter', 'Hellfire Eliminator', 'Goblin Wildfire']
    
cards = {'Goblin Bombardier': ['goblin', 'red', 'etb', 'creature'], 
         'Elvish Visionary': ['elf', 'green', 'draw', 'etb', 'creature'], 
         'Braintaster': ['zombie', 'discard', 'etb', 'black', 'creature'], 
         'Mogg Fanatic': ['goblin', 'burn', 'sacrifice', 'red', 'creature'], 
         'Crystal Shard': ['colorless', 'artifact', 'bounce'], 
         'Island': ['land', 'blue'], 
         'Goblin Char': ['instant', 'goblin', 'red', 'burn'], 
         'Goblin Dragon Hunter': ['goblin', 'creature', 'tutor', 'red'], 
         'Hellfire Eliminator': ['dragon', 'creature', 'etb'], 
         'Goblin Wildfire': ['sorcery', 'goblin', 'burn','red']    
}    

# card to card:

r1 = RuleDirect('Goblin Dragon Hunter', 'Hellfire Eliminator', 5)

# card to condition

r2 = RuleOneWay('Goblin Wildfire', ['creature', 'goblin'], 5)

# condition to condition

r3 = RuleTwoWay(['etb', 'creature'], ['bounce'], 4)
r4 = RuleOneWay('Goblin Char', ['goblin', 'creature'], 5)
r5 = RuleOneWay('Mogg Fanatic', ['goblin', 'creature'], 5)

rules = [r1, r2, r3, r4, r5]

sc = synergyChecker(rules, deck, cards)
sc.validateAllRules()

# next steps: condition in form of dict { type: creature, subtypes: [goblin]} and comparison
# for now it should be possible to make entire eynergy based on tags
# best way to go card by card then make Direct rules for non-obvious synergy and OneWay for more obvious
# Two way may be potentially dangerous (etb  + bounce, etb + blink, self-discard + reanimation, reanimation + huge)
# first 2-way should be applied then 1-way, then direct AND always if new synergy is 'stronger' 
# then it should override the weaker one
# ex. reanimation plus huge is ok but if this is durable creature or like Ulamog CH then it is much stronger

# next step: load all cards with tags and work on the rules 2- way and 1-way
```
